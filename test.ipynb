```python
# from sklearn.preprocessing import OneHotEncoder
import pandas as pd
```

## Schools

```python
va = pd.read_csv("./data/new/va/EducationDataPortal_02.08.2022_Schools.csv")
sc = pd.read_csv("./data/new/sc/EducationDataPortal_02.08.2022_Schools.csv")
nc = pd.read_csv("./data/new/nc/EducationDataPortal_02.08.2022_Schools.csv")
md = pd.read_csv("./data/new/md/EducationDataPortal_02.08.2022_Schools.csv")
tn = pd.read_csv("./data/new/tn/EducationDataPortal_02.08.2022_Schools.csv")
schools = va.append(sc.append(nc.append(md.append(tn))))

# Remove magent, charter and virtual schools
schools = schools[schools["magnet"] == "No"]
schools = schools[schools["virtual"] == "No"]
schools = schools[schools["charter"] == "No"]
schools = schools.drop(["magnet", "virtual", "charter"], axis = 1)

# Calculate percentages
schools["perc_free_lunch"] = schools["free_lunch"] / schools["enrollment"]
schools["perc_reduced_lunch"] = schools["reduced_price_lunch"] / schools["enrollment"]

# Subset only regular, graded schools 
schools = schools[schools["school_type"] == "Regular school"]
schools = schools[~schools["lowest_grade_offered"].isin(["Ungraded", "Not applicable"])]
schools = schools[~schools["highest_grade_offered"].isin(["Ungraded", "Not applicable"])]

# Replace Pre-K with 99 and Kindergarten with 0 and convert to integer
schools = schools.replace("Pre-K", 99)
schools = schools.replace("Kindergarten", 0)
schools["lowest_grade_offered"] = schools["lowest_grade_offered"].astype(int)
schools["highest_grade_offered"] = schools["highest_grade_offered"].astype(int)

# Not important column
schools = schools.drop(["year", "lea_name", "school_name", "direct_certification", "free_or_reduced_price_lunch", "title_i_eligible", "title_i_schoolwide", "free_lunch", "reduced_price_lunch", "shared_time", "enrollment", "school_type", "school_status", "seasch", "school_id"], axis = 1)

# Encode
schools["school_level"] = schools["school_level"].astype('category').cat.codes
schools["title_i_status"] = schools["title_i_status"].astype('category').cat.codes
schools["lunch_program"] = schools["lunch_program"].astype('category').cat.codes
schools["city_location"] = schools["city_location"].astype('category').cat.codes
schools["state_location"] = schools["state_location"].astype('category').cat.codes
schools["urban_centric_locale"] = schools["urban_centric_locale"].astype('category').cat.codes
schools['state_leg_district_lower'] = schools['state_leg_district_lower'].astype("category").cat.codes

print(schools.shape)

schools.head()
```

## Race

```python
va_race = pd.read_csv("./data/new/va/EducationDataPortal_02.08.2022_race.csv")
va_sex = pd.read_csv("./data/new/va/EducationDataPortal_02.08.2022_sex.csv")
va_race = va_race.pivot(index='ncessch', columns='race', values='enrollment').reset_index()
va_sex = va_sex.pivot(index='ncessch', columns='sex', values='enrollment').reset_index()
va = pd.merge(va_race, va_sex, on = "ncessch")
va = va.drop(["Total_x", "Unknown_x", "Total_y", "Unknown_y"], axis = 1)
va.head()
```

```python
nc_race = pd.read_csv("./data/new/nc/EducationDataPortal_02.08.2022_race.csv")
nc_sex = pd.read_csv("./data/new/nc/EducationDataPortal_02.08.2022_sex.csv")
nc_race = nc_race.pivot(index='ncessch', columns='race', values='enrollment').reset_index()
nc_sex = nc_sex.pivot(index='ncessch', columns='sex', values='enrollment').reset_index()
nc = pd.merge(nc_race, nc_sex, on = "ncessch")
nc = nc.drop(["Total_x", "Unknown_x", "Total_y", "Unknown_y"], axis = 1)
nc.head()
```

```python
tn_race = pd.read_csv("./data/new/tn/EducationDataPortal_02.08.2022_race.csv")
tn_sex = pd.read_csv("./data/new/tn/EducationDataPortal_02.08.2022_sex.csv")
tn_race = tn_race.pivot(index='ncessch', columns='race', values='enrollment').reset_index()
tn_sex = tn_sex.pivot(index='ncessch', columns='sex', values='enrollment').reset_index()
tn = pd.merge(tn_race, tn_sex, on = "ncessch")
tn = tn.drop(["Total_x", "Unknown_x", "Total_y", "Unknown_y"], axis = 1)
tn.head()
```

```python
sc = pd.read_csv("./data/new/sc/EducationDataPortal_02.08.2022_raceSex.csv")

sc_sex = pd.DataFrame(sc.groupby(["ncessch", "sex"])['enrollment'].sum()).reset_index()
sc_sex = sc_sex.pivot(index='ncessch', columns='sex', values='enrollment').reset_index()
sc_sex = sc_sex.drop(["Total", "Unknown"], axis = 1)

sc_race = pd.DataFrame(sc.groupby(["ncessch", "race"])['enrollment'].sum()).reset_index()
sc_race = sc_race.pivot(index='ncessch', columns='race', values='enrollment').reset_index()
sc_race = sc_race.drop(["Total", "Unknown"], axis = 1)

sc = pd.merge(sc_race, sc_sex, on = "ncessch")

sc
```

```python
md = pd.read_csv("./data/new/md/EducationDataPortal_02.08.2022_raceSex.csv")

md_sex = pd.DataFrame(md.groupby(["ncessch", "sex"])['enrollment'].sum()).reset_index()
md_sex = md_sex.pivot(index='ncessch', columns='sex', values='enrollment').reset_index()
md_sex = md_sex.drop(["Total", "Unknown"], axis = 1)

md_race = pd.DataFrame(md.groupby(["ncessch", "race"])['enrollment'].sum()).reset_index()
md_race = md_race.pivot(index='ncessch', columns='race', values='enrollment').reset_index()
md_race = md_race.drop(["Total", "Unknown"], axis = 1)

md = pd.merge(md_race, md_sex, on = "ncessch")

md
```
