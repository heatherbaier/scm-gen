```python
import geopandas as gpd
import pandas as pd
```

## Race

```python
va = pd.read_csv("./data/race/va/EducationDataPortal_02.06.2022_race.csv")
va = va[va["year"] == 2018]
va = va[['ncessch', 'state_location', 'school_name', 'race', 'enrollment']]
va.columns = ['ncessch', 'state_location', 'school_name', 'race', 'count']
va.head()
```

```python
md_sc = pd.read_csv("./data/race/md_sc/EducationDataPortal_02.06.2022_race.csv")
md_sc = md_sc[md_sc["year"] == 2018]
# print(md_sc['year'].value_counts())
md_sc = md_sc[['ncessch', 'state_location', 'school_name', 'race', 'enrollment']]
md_sc.columns = ['ncessch', 'state_location', 'school_name', 'race', 'count']
md_sc.head()
```

```python
nc = pd.read_csv("./data/race/nc/EducationDataPortal_02.06.2022_race.csv")
# print(nc['year'].value_counts())
nc = nc[nc["year"] == 2018]
nc = nc[['ncessch', 'state_location', 'school_name', 'race', 'enrollment']]
nc.columns = ['ncessch', 'state_location', 'school_name', 'race', 'count']
nc.head()
```

```python
race = va.append(md_sc.append(nc))
# race['school_id'] = race['school_name'] + race['state_location']
race
```

```python
race_meta = race[["ncessch", "state_location", "school_name"]]
race_meta.head()
```

```python
import numpy as np
```

```python
raceP = pd.merge(race_meta, race.pivot(index='ncessch', columns='race', values='count').reset_index(), on = "ncessch").drop_duplicates(subset = "ncessch")
raceP["school_name"] = raceP["school_name"].str.title().str.replace("Elem", "Elementary")
raceP = raceP.drop(["state_location", "school_name"], axis = 1)
# raceP['ncessch'] = raceP["ncessch"].astype(str).str[0:7].astype(int)
raceP.head()
```

## Funding

```python
vaF = pd.read_csv("./data/funding/VA_1819_final_January_28th_22.csv")
mdF = pd.read_csv("./data/funding/MD_1819_final_January_28th_22.csv")
scF = pd.read_csv("./data/funding/SC_1819_final_January_28th_22.csv")
ncF = pd.read_csv("./data/funding/NC_1819_final_January_28th_22.csv")

funding = vaF.append(mdF.append(scF.append(ncF)))

funding = funding[["ncesid", "distname", "schoolname", "pp_stloc_raw_MD", "pp_stloc_raw_NC", "pp_stloc_raw_VA", "pp_stloc_raw_SC"]]

funding['pp_stloc_raw_MD'] = funding['pp_stloc_raw_MD'].fillna(0)
funding['pp_stloc_raw_VA'] = funding['pp_stloc_raw_VA'].fillna(0)
funding['pp_stloc_raw_NC'] = funding['pp_stloc_raw_NC'].fillna(0)
funding['pp_stloc_raw_SC'] = funding['pp_stloc_raw_SC'].fillna(0)

funding = funding.replace("NRD", 0)
funding = funding.replace("NA", 0)

funding[['pp_stloc_raw_MD', 'pp_stloc_raw_VA', 'pp_stloc_raw_SC', 'pp_stloc_raw_NC']] = funding[['pp_stloc_raw_MD', 'pp_stloc_raw_VA', 'pp_stloc_raw_SC', 'pp_stloc_raw_NC']].astype(float)
funding["per_pupil_state"] = funding["pp_stloc_raw_MD"] + funding["pp_stloc_raw_VA"] + funding["pp_stloc_raw_SC"] + funding["pp_stloc_raw_NC"]

funding = funding[funding["per_pupil_state"] != 0]

funding = funding.rename(columns = {"ncesid": "ncessch"})

funding = funding[['ncessch', 'per_pupil_state']]

funding.head()
```

## Geography

```python
va_nc = pd.read_csv("./data/geo/va_nc/EducationDataPortal_02.06.2022_Schools.csv")
va_nc.head()
```

```python
md_sc = pd.read_csv("./data/geo/md_sc/EducationDataPortal_02.06.2022_Schools.csv")
md_sc.head()
```

```python
geo = va_nc.append(md_sc)
geo.head()
```

## Merge

```python
all_data = pd.merge(funding, raceP, on = "ncessch")
all_data = pd.merge(all_data, geo, on = "ncessch")
all_data.head()
```

```python
all_data.to_csv("./model_dataset.csv", index = False)
```
